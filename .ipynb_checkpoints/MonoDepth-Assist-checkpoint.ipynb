```python
# -------------------- Cityscapes YOLO + Depth + Error Analysis + Side-by-Side --------------------
import os, cv2, numpy as np, torch, pandas as pd
import matplotlib.pyplot as plt
from ultralytics import YOLO
from transformers import AutoImageProcessor, AutoModelForDepthEstimation

# -------------------- Config --------------------
DATASET = "Cityscapes"
USE_GT = True

CS_IMG_DIR = "Computer_vision_project/leftImg8bit_trainvaltest/leftImg8bit/val/frankfurt1"
CS_GT_DIR = "Computer_vision_project/gtFine_trainvaltest/gtFine/val/frankfurt"

YOLO_WEIGHTS = "yolov8n.pt"
OBSTACLE_CLASS_NAMES = {"person","bicycle","car","motorcycle","bus","truck","train","traffic light","stop sign","bench"}
AVG_HEIGHTS = {"person":1.7,"car":1.5,"bus":3.0,"truck":3.5,"bicycle":1.2,"motorcycle":1.2,"traffic light":3.0,"bench":0.8}

IMAGE_WIDTH = 2048
ASSUMED_HORIZONTAL_FOV_DEG = 90.0
FOCAL_LENGTH_PX = IMAGE_WIDTH / (2*np.tan(np.deg2rad(ASSUMED_HORIZONTAL_FOV_DEG/2)))

CITYSCAPES_LABELS = {24:"person",25:"rider",26:"car",27:"truck",28:"bus",
                      32:"bicycle",33:"motorcycle",20:"traffic light",23:"stop sign",11:"bench"}

# -------------------- Helper Functions --------------------
def load_image(path): return cv2.imread(path, cv2.IMREAD_COLOR)

def run_yolo(img_bgr, weights=YOLO_WEIGHTS, conf=0.25):
    model = YOLO(weights)
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    results = model.predict(source=img_rgb, conf=conf, verbose=False)[0]
    detections = []
    names = model.model.names
    for b in results.boxes:
        cls_id = int(b.cls.item())
        cls_name = names.get(cls_id,str(cls_id))
        if cls_name not in OBSTACLE_CLASS_NAMES: continue
        xyxy = b.xyxy.squeeze().cpu().numpy()
        conf_score = float(b.conf.item())
        detections.append({"cls_name":cls_name,"conf":conf_score,"xyxy":xyxy})
    return detections

def load_depth_anything(model_id="depth-anything/Depth-Anything-V2-small-hf"):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    processor = AutoImageProcessor.from_pretrained(model_id)
    model = AutoModelForDepthEstimation.from_pretrained(model_id).to(device)
    model.eval()
    return processor, model, device

def run_depth_anything(img_bgr, processor, model, device):
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    inputs = processor(images=img_rgb, return_tensors="pt").to(device)
    with torch.no_grad():
        outputs = model(**inputs)
        depth = outputs.predicted_depth.squeeze().cpu().numpy().astype(np.float32)
    return (depth - depth.min()) / (depth.max()-depth.min()+1e-8)

# -------------------- Error & IoU Functions --------------------
def median_depth_in_box(depth_map, xyxy):
    x1,y1,x2,y2 = xyxy.astype(int)
    h,w = depth_map.shape[:2]
    x1,x2 = np.clip([x1,x2],0,w-1)
    y1,y2 = np.clip([y1,y2],0,h-1)
    if x2<=x1 or y2<=y1: return np.nan
    roi = depth_map[y1:y2,x1:x2]
    return float(np.nanmedian(roi)) if roi.size>0 else np.nan

def direction_from_box(xyxy,img_w):
    cx = 0.5*(xyxy[0]+xyxy[2])
    if cx<img_w/3: return "Left"
    elif cx>2*img_w/3: return "Right"
    return "Center"

def approximate_distance(xyxy,cls_name,focal_px=FOCAL_LENGTH_PX):
    x1,y1,x2,y2 = xyxy.astype(int)
    bbox_height_px = max(1,y2-y1)
    real_height_m = AVG_HEIGHTS.get(cls_name,1.7)
    return (focal_px*real_height_m)/bbox_height_px

def compute_iou(pred_xyxy,gt_label_map,cls_name):
    x1,y1,x2,y2 = pred_xyxy.astype(int)
    x1,x2 = max(0,x1), min(gt_label_map.shape[1]-1,x2)
    y1,y2 = max(0,y1), min(gt_label_map.shape[0]-1,y2)
    if x2<=x1 or y2<=y1: return np.nan
    pred_mask = np.zeros_like(gt_label_map,dtype=np.uint8)
    pred_mask[y1:y2,x1:x2] = 1
    gt_labels = [k for k,v in CITYSCAPES_LABELS.items() if v==cls_name]
    if not gt_labels: return np.nan
    gt_mask = np.isin(gt_label_map,gt_labels).astype(np.uint8)
    intersection = np.logical_and(pred_mask,gt_mask).sum()
    union = np.logical_or(pred_mask,gt_mask).sum()
    return intersection/union if union>0 else np.nan

def compute_depth_errors(depth_map, xyxy, cls_name):
    pred_val = median_depth_in_box(depth_map, xyxy)
    if np.isnan(pred_val): return None, None
    # Approximate distance using object height
    approx_m = approximate_distance(xyxy, cls_name)
    abs_error = abs(pred_val - approx_m)
    rel_error = abs_error/approx_m if approx_m>0 else None
    return abs_error, rel_error

# -------------------- Display & Collect --------------------
def display_and_collect(df_list,img_bgr,detections,depth_map,gt_label_map=None):
    vis = img_bgr.copy()
    h,w = img_bgr.shape[:2]
    data = []
    for det in detections:
        dist_rel = median_depth_in_box(depth_map,det["xyxy"])
        if np.isnan(dist_rel): continue
        direction = direction_from_box(det["xyxy"],w)
        dist_m = approximate_distance(det["xyxy"],det["cls_name"])
        iou = compute_iou(det["xyxy"],gt_label_map,det["cls_name"]) if USE_GT and gt_label_map is not None else None
        abs_err, rel_err = compute_depth_errors(depth_map, det["xyxy"], det["cls_name"])
        x1,y1,x2,y2 = det["xyxy"].astype(int)
        cv2.rectangle(vis,(x1,y1),(x2,y2),(0,255,0),2)
        label = f"{det['cls_name']} {det['conf']:.2f} | {dist_rel:.2f} | {dist_m:.1f}m"
        (tw,th),_ = cv2.getTextSize(label,cv2.FONT_HERSHEY_SIMPLEX,0.5,1)
        cv2.rectangle(vis,(x1,max(0,y1-th-6)),(x1+tw+6,y1),(255,255,255),-1)
        cv2.putText(vis,label,(x1+3,max(0,y1-4)),cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,0,0),1)
        data.append({"class":det["cls_name"],"confidence":det["conf"],
                     "distance_rel":dist_rel,"distance_m":dist_m,
                     "direction":direction,"iou":iou,
                     "abs_err":abs_err,"rel_err":rel_err})
    df_list.append(pd.DataFrame(data))
    return vis

# -------------------- Main Pipeline --------------------
# -------------------- Full Main Loop: Cityscapes YOLO + Depth + Depth Errors --------------------
processor, depth_model, device = load_depth_anything()

# Load images and ground truth
image_files = sorted([os.path.join(CS_IMG_DIR, f) for f in os.listdir(CS_IMG_DIR) if f.endswith("_leftImg8bit.png")])
gt_files = sorted([os.path.join(CS_GT_DIR, f) for f in os.listdir(CS_GT_DIR) if f.endswith("_gtFine_labelIds.png")]) if USE_GT else [None]*len(image_files)

all_dfs = []

for idx, img_path in enumerate(image_files):
    print(f"[INFO] Processing {img_path} ...")
    img_bgr = load_image(img_path)
    h, w = img_bgr.shape[:2]
    gray_img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
    
    # YOLO detection and depth estimation
    detections = run_yolo(img_bgr)
    depth_map = run_depth_anything(img_bgr, processor, depth_model, device)
    gt_label_map = cv2.imread(gt_files[idx], cv2.IMREAD_UNCHANGED) if USE_GT else None

    # Annotate with bounding boxes + depth error overlay
    annotated = display_and_collect(all_dfs, img_bgr, detections, depth_map, gt_label_map)
    
    # Resize images
    target_w, target_h = w // 3, h // 3
    gray_resized = cv2.resize(gray_img, (target_w, target_h))
    annotated_resized = cv2.resize(annotated, (target_w, target_h))

    # Depth heatmap
    depth_norm = (depth_map - np.min(depth_map)) / (np.max(depth_map) - np.min(depth_map) + 1e-8)
    depth_colormap = cv2.applyColorMap((depth_norm * 255).astype(np.uint8), cv2.COLORMAP_JET)
    depth_resized = cv2.resize(depth_colormap, (target_w, target_h))

    # Convert grayscale to 3-channel
    gray_3ch = cv2.cvtColor(gray_resized, cv2.COLOR_GRAY2BGR)

    # Add small gap between images
    gap = 20
    gap_img = 255 * np.ones((target_h, gap, 3), dtype=np.uint8)

    # Combine side-by-side: grayscale | depth | gap | annotated
    combined = np.hstack([gray_3ch, depth_resized, gap_img, annotated_resized])

    # Display
    plt.figure(figsize=(12, 4))
    plt.imshow(cv2.cvtColor(combined, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()

    # Print table of detected objects with metrics
    if all_dfs[-1] is not None and not all_dfs[-1].empty:
        df = all_dfs[-1][["class", "distance_rel", "direction", "abs_err", "rel_err"]]
        print(f"--- Detected Objects in {os.path.basename(img_path)} ---")
        print(df.to_string(index=False))
    else:
        print(f"[INFO] No valid detections in {os.path.basename(img_path)}.")

# -------------------- Combine Results Across All Images --------------------
combined_df = pd.concat(all_dfs, ignore_index=True)
print("\n=== Summary Across All Images ===")
print("Total detections:", len(combined_df))
print("Detections per class:\n", combined_df["class"].value_counts())
if USE_GT:
    print("Mean IoU per class:\n", combined_df.groupby("class")["iou"].mean())

# -------------------- Plots --------------------
if USE_GT:
    plt.figure(figsize=(10,5))
    combined_df.groupby("class")["iou"].mean().sort_values(ascending=False).plot(kind="bar", color="skyblue")
    plt.title("Mean IoU per class")
    plt.ylabel("IoU")
    plt.ylim(0,1)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

plt.figure(figsize=(6,4))
combined_df["direction"].value_counts().plot(kind="bar", color="lightblue", edgecolor="black")
plt.title("Object Direction Distribution")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,5))
combined_df["distance_m"].hist(bins=30, color="salmon", edgecolor="black")
plt.title("Estimated Distance Distribution (m)")
plt.xlabel("Distance (m)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,5))
combined_df["abs_err"].dropna().hist(bins=30, color="lightcoral", edgecolor="black")
plt.title("Absolute Depth Error Histogram")
plt.xlabel("Absolute Error (m)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,5))
combined_df["rel_err"].dropna().hist(bins=30, color="lightgreen", edgecolor="black")
plt.title("Relative Depth Error Histogram")
plt.xlabel("Relative Error")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

```
